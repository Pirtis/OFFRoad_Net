```python
import matplotlib.pyplot as plt
import numpy as np


step, reward, dist, finish, dtp, actor_loss, critic_loss, q1_new, q2_new, log_prob = np.loadtxt("logs/reward_log.txt", delimiter=",", unpack=True)

def split_mean(arr, parts=100):
    n = len(arr)
    chunk_size = n // parts  
    
    arr = arr[:chunk_size * parts]
 
    return arr.reshape(parts, chunk_size).mean(axis=1)

step_2 = step
step       = split_mean(step)
reward     = split_mean(reward)
dist       = split_mean(dist)
# finish     = split_mean(finish)
# dtp        = split_mean(dtp)
actor_loss = split_mean(actor_loss)
critic_loss= split_mean(critic_loss)
q1_new     = split_mean(q1_new)
q2_new     = split_mean(q2_new)
log_prob   = split_mean(log_prob)
# action     = split_mean(action)
# action_2   = split_mean(action_2)

plt.figure(figsize=(10, 5))
plt.plot(step, reward, label="Total Reward per Episode")
plt.xlabel("step")
plt.ylabel("Total Reward")
plt.title("Training Progress")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(step, dist, label="Total dist per Episode")
plt.xlabel("step")
plt.ylabel("Dist")
plt.title("Training Progress")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(step_2, finish, label="Total step per Episode")
plt.xlabel("step")
plt.ylabel("finish")
plt.title("Training Progress")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(step_2, dtp, label="Check finish per Episode")
plt.xlabel("step")
plt.ylabel("dtp")
plt.title("Training Progress")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(step, actor_loss, label="Check dtp per Episode")
plt.xlabel("step")
plt.ylabel("actor_loss")
plt.title("Training Progress")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(step, critic_loss, label="Epsilon per Episode")
plt.xlabel("step")
plt.ylabel("critic_loss")
plt.title("Training Progress")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(step, q1_new, label="Epsilon per Episode")
plt.xlabel("step")
plt.ylabel("q1_new")
plt.title("Training Progress")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(step, q2_new, label="Epsilon per Episode")
plt.xlabel("step")
plt.ylabel("q2_new")
plt.title("Training Progress")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(step, log_prob, label="Epsilon per Episode")
plt.xlabel("step")
plt.ylabel("log_prob")
plt.title("Training Progress")
plt.grid(True)
plt.legend()
plt.show()
```
